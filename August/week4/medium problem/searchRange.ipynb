```python
class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        '''
            review binary search
        '''
        left = 0
        right = len(nums) - 1
        
        if len(nums) == 1:
            if nums[0] == target:
                return [0,0]
            else:
                return [-1,-1]

        while(left <= right):
            middle = int( (left + right)/2)
            print('middle:%d' %(middle))
            if nums[middle] == target:
                if middle == len(nums) - 1:
                    return [middle, middle]
                elif middle == 0:
                    if nums[middle + 1] == target:
                        return [middle, middle + 1]
                    else:
                        return [middle, middle]                    
                else:
                    if nums[middle - 1] != target and nums[middle + 1] != target:
                        return [middle, middle]
                    else:
                        head = tail = middle
                        #find head of target
                        while (1):
                            print('head:%d' %(head) )
                            if head == 0:
                                break
                            else:
                                if nums[head - 1] == target:
                                    head = head - 1
                                else:
                                    break
                        #find tail of target        
                        while (1):
                            print('tail:%d' %(tail) )
                            if tail == len(nums) - 1:
                                break
                            else:
                                if nums[tail + 1] == target:
                                    tail = tail + 1
                                else:
                                    break

                        return [head,tail]
                        
            elif nums[middle] > target:
                right = middle - 1
            else:
                left = middle + 1
                
        return [-1, -1]
```
